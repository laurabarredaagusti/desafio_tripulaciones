```python
def calc_advanced(precio_1,precio_2, gasto_1, gasto_2, n_meses):
    '''     
    Calculate the crossing point where the most expensive home appliance starts to be cheaper if its 
    consumption is less.

    Args:
        precio_1 = price of the first home appliance. (€)
        precio_2 = price of the second home appliance. (€)
        gasto_1 = cost of the first home appliance in a month. (€)
        gasto_2 = cost of the second home appliance in a month. (€)
        n_meses = number of months home appliances are used in a year.       

    Returns:
        Points to graph the consumption of appliances over the years.

    '''    
    lista = []
    
    punto_ini_1 = (0, precio_1) # punto inicial del aparato 1
    punto_ini_2 = (0, precio_2) # punto inicial del aparato 2
    
    lista.append(punto_ini_1)
    lista.append(punto_ini_2)    
    
    
    # else:
    if precio_1!=precio_2 and gasto_1!=gasto_2 and ((precio_1>precio_2 and gasto_1<gasto_2) or (precio_1<precio_2 and gasto_1>gasto_2) ):
        
        tiempo_coinciden = (precio_1 - precio_2)* (1/(gasto_2 - gasto_1)) * (1/n_meses)

        punto_corte = (abs(tiempo_coinciden), precio_2 + gasto_2 * tiempo_coinciden * n_meses ) 

        lista.append(punto_corte)

        # SI SE CRUZAN EN MÁS DE 15 AÑOS
        if (abs(tiempo_coinciden))>15:
            punto_fin_1 = (15 ,precio_1 + gasto_1 * (15) * n_meses) # punto final del aparato 1
            punto_fin_2 = (15, precio_2 + gasto_2 * (15) * n_meses) # punto final del aparato 2
        
            lista.append(punto_fin_1)
            lista.append(punto_fin_2)
        
            return lista


        else:    
            
            punto_fin_1 = (tiempo_coinciden *1.3 ,precio_1 + gasto_1 * (tiempo_coinciden *1.3) * n_meses) # punto final del aparato 1
            punto_fin_2 = (tiempo_coinciden *1.3, precio_2 + gasto_2 * (tiempo_coinciden *1.3) * n_meses) # punto final del aparato 2
        
            lista.append(punto_fin_1)
            lista.append(punto_fin_2)
        
            return lista

    else:
        punto_corte= (0,0)  #  NO EXITE. VALOR (0,0) por defecto. SE PUEDE CAMBIAR A CUALQUIER OTRO
        # EN CASO DE MAYOR PRECIO, MAS GASTO, el punto de corte es negativo, pero no nos interesa

        lista.append(punto_corte)


        punto_fin_1 = 5,(precio_1 + gasto_1 * (5) * n_meses) 
        punto_fin_2 = 5, (precio_2 + gasto_2 * (5) * n_meses) 

        lista.append(punto_fin_1)
        lista.append(punto_fin_2)

        return lista
```
