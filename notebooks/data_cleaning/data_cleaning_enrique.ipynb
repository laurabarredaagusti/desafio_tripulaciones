```python
df_calentador_agua = pd.read_csv('../csvs_categorias(Columnas Bien)/7_water_heaters.csv',index_col=0)
df_calentador_agua.head()
```

```python
df_calentador_agua.info()
```

```python
df_calentador_agua['Other'].unique()
```

```python
df_calentador_agua.info()
```

```python
df_calentador_agua['Consumption'] = df_calentador_agua['Consumption'].str.replace(' ', '')
```

```python
df_calentador_agua['Consumption'] = df_calentador_agua['Consumption'].astype(float)
```

```python
df_calentador_agua['Consumption'].unique()
```

```python
df_calentador_agua[df_calentador_agua['Consumption']<400]
```

```python
df_calentador_agua[df_calentador_agua['Consumption']==365]
```

```python
df_calentador_agua['Consumption'] = df_calentador_agua['Consumption']/365
```

```python
import numpy as np
```

```python
df_calentador_agua['Consumption_M'] = np.where( df_calentador_agua['Other'] == ' M', df_calentador_agua['Consumption']/1.67, '0')
df_calentador_agua['Consumption_L'] = np.where( df_calentador_agua['Other'] == ' L', df_calentador_agua['Consumption']/1.75, '0')
df_calentador_agua['Consumption_XL'] = np.where( df_calentador_agua['Other'] == ' XL', df_calentador_agua['Consumption']/2.25, '0')
df_calentador_agua['Consumption_XXL'] = np.where( df_calentador_agua['Other'] == ' XXL', df_calentador_agua['Consumption']/2.75, '0')
```

```python
df_calentador_agua['Consumption_f'] = df_calentador_agua['Consumption_M'] + df_calentador_agua['Consumption_L']+df_calentador_agua['Consumption_XL'] +df_calentador_agua['Consumption_XXL']
```

```python
df_calentador_agua['Consumption_f'] = df_calentador_agua['Consumption_f'].astype(float)
```

```python
df_calentador_agua.drop(columns=['Consumption_M', 'Consumption_L', 'Consumption_XL', 'Consumption_XXL'], inplace=True)
```

```python
df_calentador_agua
```

```python
df_calentador_agua.drop(columns=['Consumption'], inplace=True)
```

```python
df_calentador_agua.rename(columns={'Consumption_f': 'Consumption'}, inplace=True)
```

```python
df_calentador_agua.to_csv('7_water_heaters_c.csv')
```
